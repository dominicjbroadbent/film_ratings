## Load modules

```python
import pandas as pd
import numpy as np
```

## Rotten Tomatoes

```python
# Load base dataset
rt = pd.read_csv('rt_ratings.csv')
rt
```

```python
# Extract relevant columns
rt = rt[['Title', 'Year', 'Rotten Tomatoes','Netflix', 'Hulu', 'Prime Video', 'Disney+']]
rt
```

```python
 # Add the year to the title and drop the year column
rt['Title'] = rt['Title'] + rt['Year'].astype(str)
rt = rt.drop(columns = 'Year')
rt
```

```python
# Remove punctation, whitespace and set to lowercase
rt['Title'] = rt['Title'].str.replace('[^\w\s]', '', regex = True).str.lower().str.replace(' ', '')
rt
```

## IMDB

```python
# Read in rating dataset
imdb_ratings = pd.read_csv('imdb_ratings.tsv', sep = '\t')
imdb_ratings
```

```python
# Read in titles dataset
imdb_titles = pd.read_csv('imdb_titles.tsv', sep = '\t', low_memory = False)
imdb_titles
```

```python
# Drop irrelevant columns and filter just the films
imdb_titles = imdb_titles.drop(columns = ['endYear', 'primaryTitle', 'isAdult'])
imdb_titles = imdb_titles[imdb_titles['titleType'] == 'movie'].drop(columns = 'titleType')
imdb_titles
```

```python
# Merge the titles and ratings dataset according to IMDB's id system
imdb = pd.merge(imdb_titles, imdb_ratings, on = 'tconst')
imdb
```

```python
# Combine the release year with the title and drop the year column
imdb['originalTitle'] = imdb['originalTitle'] + imdb['startYear'].astype(str)
imdb = imdb.drop(columns = 'startYear')
imdb
```

```python
# Remove punctation, whitespace and set to lowercase
imdb['originalTitle'] = imdb['originalTitle'].str.replace('[^\w\s]', '', regex = True).str.lower().str.replace(' ', '')
imdb
```

## Combine dataset

```python
# Merge the datasets based on the titles, we lose lots of films here as the rotten tomatoes dataset is significantly more limited than IMDB's
imdb_rt = pd.merge(imdb, rt, left_on = 'originalTitle', right_on = 'Title')
imdb_rt
```

```python
# Drop any duplicate films that the title-year standardistation has not been able to catch
# Also drop the duplicate title column
imdb_rt = imdb_rt.drop_duplicates(subset = ['Title'], keep = False).drop(columns = 'originalTitle')
imdb_rt
```

```python
# Convert the fraction strings to floats, and scale them up to be out of 10
imdb_rt['Rotten Tomatoes'] = imdb_rt['Rotten Tomatoes'].str.split('/').apply(lambda x: float(x[0]) / float(x[1]) * 10)
imdb_rt
```

```python
# Drop unique id
imdb_rt = imdb_rt.drop(columns = 'tconst')
imdb_rt
```

```python
# Reorganise columns
imdb_rt = imdb_rt[['Title','genres', 'Netflix', 'Hulu', 'Prime Video', 'Disney+', 'runtimeMinutes', 'numVotes', 'averageRating', 'Rotten Tomatoes']]
imdb_rt
```

```python
# Name columns nicely
col_names = ['title','genre','netflix', 'hulu', 'prime', 'disney', 'runtime', 'num_votes', 'imdb_rating', 'rt_rating']
imdb_rt.columns = col_names
imdb_rt
```

```python
#Reset the index
imdb_rt = imdb_rt.reset_index(drop = True)
imdb_rt
```

### Expand genre info

```python
# Split the genre strings into a list of lists
genre_list = [item.split(',') for item in imdb_rt['genre'].values]
genre_list[0:5]
```

```python
# Get a list of all the unique genres
unique_genres = list(set(np.array([item for sublist in genre_list for item in sublist])))
unique_genres[0:5]
```

```python
# Make a dictionary to map genre to column index
genre_to_index = {genre: index for genre, index in zip(unique_genres, range(len(unique_genres)))} 
genre_to_index
```

```python
# Initalise array to hold genre data
genre_array = np.zeros( ( len(genre_list), len(unique_genres) ) )
```

```python
# Loop over each of the genres of each film, and put a 1 in the corresponding columns
for i, genres in enumerate(genre_list):
    for genre in genres:
        col_index = genre_to_index[genre]
        genre_array[i, col_index] = 1

genre_array
```

```python
# Make a dataframe out of the numpy array
genre_df = pd.DataFrame(genre_array)

# Add column names
unique_genres = [genre.lower() for genre in unique_genres]
genre_df.columns = unique_genres
genre_df
```

```python
# Replace the genre column with this encoded data 
imdb_rt = imdb_rt.drop(columns = ['genre'])
imdb_rt = pd.concat([imdb_rt, genre_df], axis = 1)
imdb_rt
```

```python
# Move the imdb rating to the end for sanity
cols = list(imdb_rt.columns.values) 
cols.pop(cols.index('imdb_rating')) 
imdb_rt = imdb_rt[cols + ['imdb_rating']] 
```
